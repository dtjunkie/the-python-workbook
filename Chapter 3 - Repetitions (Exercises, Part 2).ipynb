## Exercise 71 - Approximate $ π $

The value of $π $ can be approximated by the following infinite series:

$π \approx 3 + \frac{4}{2 * 3 * 4} - \frac{4}{4 * 5 * 6} + \frac{4}{6 * 7 * 8} - \frac{4}{8 * 9 * 10} + \frac{4}{10 * 11 * 12} - ... $

Write a program that displays 15 approximations of $π $. The first approximation should make use of only the first term from the infinite series. Each additional
approximation displayed by your program should include one more term in the series,
making it a better approximation of π than any of the approximations displayed previously (23 lines).

```python
# Created a variable with the initial value of 3
pi = 3

# Created a variable for the denominator with an initial value of 2
d = 2

# Created a variable for the number of approximation with an initial value of 1
n = 1

# Prints all 15 approximations of π
print("The 15 approximations of \u03C0 are: ")
while n <= 15:
    if n == 1:
        print(pi)
        n += 1
    else:
        term = 4 / (d * (d + 1) * (d + 2))
        if n % 2 == 0:
            pi += term
        else:
            pi -= term
        print(pi)
        d += 2
        n += 1            
```

## Exercise 72 - Fizz-Buzz

Fizz-Buzz is a game that is sometimes played by children to help them learn about
division. The players are commonly arranged in a circle so that the game can progress
from player to player continually. The starting player begins by saying one, and then
play passes to the player to the left. Each subsequent player is responsible for the
next integer in sequence before play passes to the following player. On a player’s
turn they must either say their number or one of following substitutions:

- If the player’s number is divisible by 3 then the player says fizz instead of their number.
- If the player’s number is divisible by 5 then the player says buzz instead of their number.

A player must say both fizz and buzz for numbers that are divisible by both 3
and 5. Any player that fails to perform the correct substitution or hesitates before
answering is eliminated from the game. The last player remaining is the winner.
Write a program that displays the answers for the first 100 numbers in the Fizz-Buzz game. Each answer should be displayed on its own line (17 lines).

```python
# Displays the answers of the first 100 numbers in the Fizz-Buzz Game
for i in range(1,101):
    # Checking if the range number is divisible by 3
    if i % 3 == 0:
        # Checking if the range number is also divisible by 5
        # If the range number is both divisible by 3 and 5, it will print "Fizz-Buzz".
        if i % 5 == 0:
            print("Fizz-Buzz")
        # Else, it will print "Fizz".
        else:
            print("Fizz")
    #Checking if the range number is divisible by 5
    else:
        # If true, then it will display "Buzz"
        if i % 5 == 0:
            print("Buzz")
        # Else, it will display the range number
        else:
            print(i)
```

## Exercise 73 -  Caesar Cipher

</p> One of the first known examples of encryption was used by Julius Caesar. Caesar
needed to provide written instructions to his generals, but he didn’t want his enemies
to learn his plans if the message slipped into their hands. As a result, he developed
what later became known as the Caesar cipher.<p>
</p>The idea behind this cipher is simple (and as such, it provides no protection against
modern code breaking techniques). Each letter in the original message is shifted by
3 places. As a result, A becomes D, B becomes E, C becomes F, D becomes G, etc.
The last three letters in the alphabet are wrapped around to the beginning: X becomes
A, Y becomes B and Z becomes C. Non-letter characters are not modified by the
cipher.<p>
</p>Write a program that implements a Caesar cipher. Allow the user to supply the
message and the shift amount, and then display the shifted message. Ensure that
your program encodes both uppercase and lowercase letters. Your program should
also support negative shift values so that it can be used both to encode messages and
decode messages (35 lines).<p>

```python
# Displays the title of the program
print("CEASAR CIPHER")

'''
This program shifts every letter of the user's message by a value
provided by the user.
'''

# Asks the user to input a message
message = input("Enter a message: ")

# Asks the user to input the shift value
shift_value = int(input("Enter the shift value: "))

# Created a variable with an empty string where we will put all the encoded letters
encrypted_message = ""

# Encrypts the letters up to the last letter
for char in message:
    # Process the lower case letter by determining its position
    # in the alphabet(0-25), shifts the position, and add the new
    # character to the variable encrypted_message
    if "a" <= char <= "z":
        pos = ord(char) - ord("a")
        new_pos = (pos + shift_value) % 26
        new_char = chr(new_pos + ord("a"))
        encrypted_message += new_char
    # Process the upper case letter by determining its position
    # in the alphabet(0-25), shifts the position, and add the new
    # character to the variable encrypted_message
    elif "A" <= char <= "Z":
        pos = ord(char) - ord("A")
        new_pos = (pos + shift_value) % 26
        new_char = chr(new_pos + ord("A"))
        encrypted_message += new_char
    # If it is not an alphabet, it will be just added to the
    # variable encrypted_message
    else:
        encrypted_message += char
        
# Displays the encrypted message
print("Encrypted Message: ", encrypted_message)
```

## Exercise 74 -  Square Root

<p>Write a program that implements Newton’s method to compute and display the
square root of a number, x, entered by the user. The algorithm for Newton’s method
follows:</p>

> Read $x$ from the user
<br> Initialize $guess$ to $x/2$
<br> **While** $guess$ is not good enough **do**
<br> &emsp;Update $guess$ to be the average of $guess$ and $x/guess$

<p>When this algorithm completes, $guess$ contains an approximation of the square
root of $x$. The quality of the approximation depends on how you define “good
enough”. In the author’s solution, $guess$ was considered good enough when the
absolute value of the difference between $guess$ ∗ $guess$ and $x$ was less than or equal
to $10^{−12}$ (14 lines). </p>

```python
print("Newton's Approximation Method For Square Root")

'''
This program computes and display the square root
of the number the user entered.
'''

# Reads the entered number by the user
x = float(input("Enter a number: "))

# Creates a variable "guess" with an initial value of x/2
guess = x/2

# Creates a variable for the no. of approximations with an initial value of zero
n = 1

# Calculates and displays the approximation until it satisfies the "while" condition.
while (abs(guess**2 - x) != 10**-12 or abs(guess**2 - x) > 10**-12) and n != 31:
    guess = (guess + (x/guess))/2
    print("Approximation No. %2d" % (n)  + ": %18.12f" % (guess))
    n += 1
```

## Exercise 75 - Is a String a Palindrome?

<p>A string is a palindrome if it is identical forward and backward. For example “anna”,
“civic”, “level” and “hannah” are all examples of palindromic words.Write a program
that reads a string from the user and uses a loop to determine whether or not it is a
palindrome. Display the result, including a meaningful output message (Solved, 26 Lines)</p>.

> Aibohphobia is the extreme or irrational fear of palindromes. This word was
constructed by prepending the -phobia suffix with it’s reverse, resulting in
a palindrome. Ailihphilia is the fondness for or love of palindromes. It was
constructed in the same manner from the -philia suffix.

```python
print("PALINDROME STRING CHECKER")

'''
The checks the string if it is a palindrome.
'''
# Reads the input string of the user
string = input("Enter a word: ")

# Converts the alphabet characters to lower case
string = string.lower()

# Tracks index "i" and "j" of the string
i = 0
j = len(string) - 1

# Checks if the characters of index i and j are identical
# as long as i is still less than j
while i < j:
    # If the string of index "i" is identical with index "j",
    # it will move to the next set of characters for another
    # comparison.
    if string[i] == string[j]:
        i += 1
        j -= 1
    # Else, it will display a message that the string is not
    # a palindrome.
    else:
        print("This string is not a palindrome.")
        break
# If each sets of character are identical,
# it will display a message saying that the string is a palindrome.
else:
    print("This string is a palindrome.")
```

while determining whether or not a string is a palindrome. For an additional challenge,

```python
# Imports the punctuation function from the string module
from string import punctuation

print("PALINDROME STRING CHECKER")

'''
The checks the string if it is a palindrome.
'''
# Reads the input string of the user
string = input("Enter a word: ")

# Strips all punctuation marks from the string
punct = punctuation

for char in string:
    if char in punct:
        string = string.replace(char, " ")

# Removes all spaces from the input string
string = string.replace(" ", "")

# Converts the alphabet characters to lower case
string = string.lower()

# Tracks index "i" and "j" of the string
i = 0
j = len(string) - 1

# Checks if the characters of index i and j are identical
# as long as i is still less than j
while i < j:
    # If the string of index "i" is identical with index "j",
    # it will move to the next set of characters for another
    # comparison.
    if string[i] == string[j]:
        i += 1
        j -= 1
    # Else, it will display a message that the string is not
    # a palindrome.
    else:
        print("This string is not a palindrome.")
        break
# If each sets of character are identical,
# it will display a message saying that the string is a palindrome.
else:
    print("This string is a palindrome.")
```

## Exercise 78 -  The Collatz Conjecture

Start with any positive integer as the only term in the sequence
>**While** the last term in the sequence is not equal to 1 **do**
<br>&emsp;**If** the last term is even **then**
<br>&emsp;&emsp;**Add** another term to the sequence by dividing the last term by 2 using
<br>&emsp;&emsp;floor division
<br>&emsp;**Else**
<br>&emsp;&emsp;Add another term to the sequence by multiplying the last term by 3 and
<br>&emsp;&emsp;adding 1
          
The Collatz conjecture states that this sequence will eventually end with one when
it begins with any positive integer. Although this conjecture has never been proved,
it appears to be true.
Create a program that reads an integer, n, from the user and reports all of the
values in the sequence starting with n and ending with one. Your program should
allow the user to continue entering new n values (and your program should continue
displaying the sequences) until the user enters a value for n that is less than or equal
to zero. (22 lines)

> The Collatz conjecture is an example of an open problem in mathematics.
While many people have tried to prove that it is true, no one has been able to
do so. Information on other open problems in mathematics can be found on
the Internet.

```python
# Displays the title of the program
print(f"{'THE COLLATZ CONJECTURE CALCULATOR': ^46}")

'''
Displays a sequence of positive integers generated
by dividing the last term by 2 if it is even and 
tripling the last term and add 1 if it is odd. The
sequence ends if the last term is 1.
'''

# Asks a number from the user
n = int(input("Enter a number: "))

# The program terminates if the user enters the number 0
while n > 0:
    n_values = [] # This is where we will append all the numbers
    n_values.append(n) # Adds the first number into the list
    # The program will continue to calculate until the last term
    # becomes 1. Afterwards, it will ask the user to enter another number.
    # The program terminates if the user enters the number 0.
    while n_values[len(n_values) - 1] != 1 and n > 0:
        # If the last term is even, it will halved before
        # adding it to the list
        if n_values[len(n_values) - 1] % 2 == 0:
            n = n_values[len(n_values) - 1] // 2
            n_values.append(n)
        # Else, it will be tripled and add 1 before
        # adding it to the list.
        else:
            n = (3 * n_values[len(n_values) - 1]) + 1 
            n_values.append(n)
    # Displays all the values
    print(n_values)
    # Asks the user to enter another number
    n = int(input("Enter a number: "))  
```

## Exercise 79 -  Greatest Common Divisor

The greatest common divisor of two positive integers, $n$ and $m$, is the largest number,
$d$, which divides evenly into both $n$ and $m$. There are several algorithms that can be
used to solve this problem, including:

&emsp;&emsp;Initialize $d$ to the smaller of $m$ and $n$.
<br>&emsp;&emsp;**While** $d$ does not evenly divide $m$ or $d$ does not evenly divide $n$ do
<br>&emsp;&emsp;&emsp;Decrease the value of $d$ by 1
<br>&emsp;&emsp;Report $d$ as the greatest common divisor of $n$ and $m$

Write a program that reads two positive integers from the user and uses this algorithm
to determine and report their greatest common divisor. (17 lines)

```python
# Displays the title of the program
print(f"{'GREATEST COMMON DIVISOR': ^46}")

'''
Displays the greatest common divisor(GCD) of two numbers
'''

# Asks the user to input two numbers
m = int(input("Input a number: "))
n = int(input("Input another number: "))

# Initializes d to the smaller of m and n
d = min(m, n)

# When m or n is not evenly divide by d,
# d will be decreased by 1
while m % d != 0 or n % d != 0:
    d -= 1

# Reports the greatest common divisor
print("%d is the greatest common divisor of %d and %d" % (d, n, m))
```

## Exercise 80 - Prime Factors

The prime factorization of an integer, $n$, can be determined using the following steps:

&emsp;&emsp;Initialize $factor$ to 2
<br>&emsp;&emsp;**While** $factor$ is less than or equal to $n$ **do**
<br>&emsp;&emsp;&emsp;**If** $n$ is evenly divisible by $factor$ **then**
<br>&emsp;&emsp;&emsp;&emsp;Conclude that $factor$ is a factor of $n$
<br>&emsp;&emsp;&emsp;&emsp;Divide $n$ by factor using floor division
<br>&emsp;&emsp;&emsp;**Else**
<br>&emsp;&emsp;&emsp;&emsp;Increase $factor$ by 1

Write a program that reads an integer from the user. If the value entered by the
user is less than 2 then your program should display an appropriate error message.
Otherwise your program should display the prime numbers that can be multiplied
together to compute $n$, with one factor appearing on each line. For example:

`Enter an integer (2 or greater): 72
The prime factors of 72 are:
2
2
2
3
3`

(27 lines)

```python
# Displays the title of the program
print(f"{'PRIME FACTOR': ^46}")

'''
Displays the prime factors of n
'''

#Asks the user to enter a number greater than or equal to 2
n = int(input("Enter a number (2 or greater): "))

# If the user enters a number less than zero, an error message
# will appear and the program will terminate
if n < 2:
    print("InputError: Input must be greater than or equal to 2.")
# Else, it will proceed with calculation
else:
    print("The prime factors of %d are: " % n)
    factor = 2 # Initialized variable factor to 2
    # The program will continue to execute as long
    # as the factor is less than or equal to n
    while factor <= n:
        # If the n is evenly divided by the factor,
        # it will print the factor and the n will be
        # divided by factor. This will be the new value
        # for the next calculation.
        if n % factor == 0:
            print(factor)
            n //= factor
        # Else, factor will be added by 1
        else:
            factor += 1
```
