## Exercise 35 -  Even or Odd?

Write a program that reads an integer from the user. Then your program should
display a message indicating whether the integer is even or odd (13 lines).

```python
# Asks the user to enter a number
num = int(input("Ënter a number: "))

# Determines if the input number is divisible by 2
div_2 = num % 2

# Decides which statement is to be printed based on output of div_2
if div_2 == 0:
    print(num,"is an even number.")
if div_2 == 1:
    print (num,"is an odd number.")
```

## Exercise 36 - Dog Years

It is commonly said that one human year is equivalent to 7 dog years. However this
simple conversion fails to recognize that dogs reach adulthood in approximately two
years. As a result, some people believe that it is better to count each of the first two
human years as 10.5 dog years, and then count each additional human year as 4 dog
years.
Write a program that implements the conversion from human years to dog years
described in the previous paragraph. Ensure that your program works correctly for
conversions of less than two human years and for conversions of two or more human
years. Your program should display an appropriate error message if the user enters
a negative number (22 lines).

```python
# Asks the user to enter the no. of human years
human_year = int(input("Enter the number of human years: "))

# Decides which action to be done and statement to be printed based on the user input
if human_year < 0:
    print("Value Error:Invalid value")
elif human_year == 0:
    print(human_year,"human year is equal to 0 dog years.")
elif human_year == 1:
    print(human_year,"human year is equal to 10.5 dog years.")
else:
    base_dog_year = 10.5
    add_human_year = human_year - 2
    add_dog_year = add_human_year * 4
    total_dog_year = base_dog_year + add_dog_year
    print(human_year,"human years is equal to",total_dog_year,"dog_years")
    
    
```

## Exercise 37 - Vowel or Consonant

In this exercise you will create a program that reads a letter of the alphabet from the
user. If the user enters a, e, i, o or u then your program should display a message
indicating that the entered letter is a vowel. If the user enters y then your program
should display a message indicating that sometimes y is a vowel, and sometimes y is
a consonant. Otherwise your program should display a message indicating that the
letter is a consonant(16 lines).

```python
# Asks the user to enter a letter
letter = input("Enter a letter: ")

# Decides which statement is to be printed based on the input
if letter == "a" or letter == "e" or letter == "i" or letter == "o" or letter == "u":
    print(letter,"is a vowel.")
elif letter == "y":
    print(letter,"is either a vowel or a consonant.")
else:
    print(letter,"is a consonant.")
```

## Exercise 38 - Name That Shape

Write a program that determines the name of a shape from its number of sides. Read
the number of sides from the user and then report the appropriate name as part of
a meaningful message. Your program should support shapes with anywhere from 3
up to (and including) 10 sides. If a number of sides outside of this range is entered
then your program should display an appropriate error message(31 lines).

```python
# Asks the user to enter the no. of sides
sides = int(input("Enter the number of sides: "))

# Decides which statement is to be printed based on the no. of sides
if sides < 3 or sides > 10:
    print("Value Error:Number of sides should more than 2 or less than 10.")
else:
    if sides == 3:
        print("This is a triangle.")
    elif sides == 4:
        print("This is a quadrilateral.")
    elif sides == 5:
        print("This is a pentagon.")
    elif sides == 6:
        print("This is a hexagon.")
    elif sides == 7:
        print("This is a heptagon.")
    elif sides == 8:
        print("This is a octagon.")
    elif sides == 9:
        print("This is a nonagon.")
    else:
        print("This is a decagon.")   
```

## Exercise 39 -  Month Name to Number of Days

The length of a month varies from 28 to 31 days. In this exercise you will create
a program that reads the name of a month from the user as a string. Then your
program should display the number of days in that month. Display “28 or 29 days”
for February so that leap years are addressed(18 lines).

```python
# Asks the user to input the month name
month_name = input("Enter the month name: ")

# Decides which statement is to be displayed based on the input
if month_name == "February":
    print(month_name,"has 28 or 29 days.")
elif month_name == "April" or month_name == "June" or month_name == "September" or month_name == "November":
    print(month_name,"has 30 days.")
else:
    print(month_name,"has 31 days.")
```

## Exercise 40 -  Sound Levels

The following table lists the sound level in decibels for several common noises.

enters a decibel level that matches one of the noises in the table then your program
should display a message containing only that noise. If the user enters a number
of decibels between the noises listed then your program should display a message
indicating which noises the value is between. Ensure that your program also generates
reasonable output for a value smaller than the quietest noise in the table, and for a
value larger than the loudest noise in the table(30 lines).

```python
# Asks the user to input the sound level
dB = int(input("Enter the sound level(in dB): "))

# Decides which statement is to be printed based on the input
if dB < 40:
    print("The sound level is less than the sound level of the noise generated by a quiet room.")
elif dB == 40:
    print("The sound level is the same as the sound level of the noise generated by a quiet room.")
elif dB > 40 and dB < 70:
    print("The sound level is between the sound level of the noise generated by a quiet room and the noise generated by an alarm clock.")
elif dB == 70:
    print("The sound level is the same as the sound level of the noise generated by an alarm clock.")
elif dB > 70 and dB < 106:
    print("The sound level is between the sound level of the noise generated by an alarm clock and the noise generated by a gas lawnmower.")
elif dB == 106:
    print("The sound level is the same as the sound level of the noise generated by an gas lawnmower.")
elif dB > 106 and dB < 130:
    print("The sound level is between the sound level of the noise generated by a gas lawnmower and the noise generated by a jackhammer.")
elif dB == 130:
    print("The sound level is the same as the sound level of the noise generated by a jackhammer.")
elif dB > 130:
    print("The sound level exceed the sound level of the noise generated by a jackhammer.")
```
