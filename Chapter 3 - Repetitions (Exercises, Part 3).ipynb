## Exercise 81 -  Binary to Decimal

Write a program that converts a binary (base 2) number to decimal (base 10). Your
program should begin by reading the binary number from the user as a string. Then
it should compute the equivalent decimal number by processing each digit in the
binary number. Finally, your program should display the equivalent decimal number
with an appropriate message. (18 Lines)

```python
# Displays the title of the program
print(f"{'BINARY TO DECIMAL': ^46}")

'''
Converts binary string to decimal number
'''
# Asks the user to input a binary number
bin_2 = input("Enter a binary number: ")

# Initialize n to 0
# To be used as exponent of 2
n = 0

# Initialize dec_10 to zero
dec_10 = 0

# Converts each string to integer
# Multiply them by 2 raised to the power of n - i
# Adding them to dec_10
for i in range(1,len(bin_2) + 1):
    add = int(bin_2[-i]) * (2 ** n)
    dec_10 += add
    n += 1
    
# Displays the decimal equvalent of the input binary number
print("The decimal equivalent is: ", dec_10)
```

## Exercise 82: Decimal to Binary

Write a program that converts a decimal (base 10) number to binary (base 2). Read the
decimal number from the user as an integer and then use the division algorithm shown
below to perform the conversion. When the algorithm completes, result contains the
binary representation of the number. Display the result, along with an appropriate
message. (27 lines)

```python
# Displays the title of the program
print(f"{'DECIMAL TO BINARY': ^46}")

'''
Converts decimal number to binary string
'''
# Asks the user to input an integer
dec_10 = int(input("Enter an integer: "))

# Created an variable w/ empty string for the
# bits converted into string
bin_2_rev = ''

# Gets the remainder of the dec_10 divided to 2
# Appends the them to the variable w/ an empty string
# Reverses the string
# Divides the input to 2
# The process continues until dec_10 is equal to 0
while dec_10 != 0:
    rem = dec_10 % 2
    bin_2_rev += str(rem)
    bin_2 = bin_2_rev[::-1]
    dec_10 //= 2
    
# Displays the binary equivalent of input decimal number
print("The binary equivalent is: ", bin_2)
```

## Exercise 83 -  Maximum Integer

<p>This exercise examines the process of identifying the maximum value in a collection
of integers. Each of the integers will be randomly selected from the numbers between
1 and 100. The collection of integers may contain duplicate values, and some of the
integers between 1 and 100 may not be present.</p>
<p>Take a moment and think about how you would solve this problem on paper. Many
people would check each integer in sequence and ask themself if the number that
they are currently considering is larger than the largest number that they have seen
previously. If it is, then they forget the previous maximum number and remember
the current number as the new maximum number. This is a reasonable approach,
and will result in the correct answer when the process is performed carefully. If you
were performing this task, how many times would you expect to need to update the
maximum value and remember a new number?</p>
<p>While we can answer the question posed at the end of the previous paragraph
using probability theory, we are going to explore it by simulating the situation. Create a program that begins by selecting a random integer between 1 and 100. Save this
integer as the maximum number encountered so far. After the initial integer has been
selected, generate 99 additional random integers between 1 and 100. Check each
integer as it is generated to see if it is larger than the maximum number encountered
so far. If it is then your program should update the maximum number encountered
and count the fact that you performed an update. Display each integer after you generate it. Include a notation with those integers which represent a new maximum.</p>
<p>After you have displayed 100 integers your program should display the maximum value encountered, along with the number of times the maximum value
was updated during the process. Partial output for the program is shown below,
with… representing the remaining integers that your program will display. Run your
program several times. Is the number of updates performed on the maximum value
what you expected?</p>


`30
74 <== Update
58
17
40
37
13
34
46
52
80 <== Update
37
97 <== Update
45
55
73
...
The maximum value found was 100
The maximum value was updated 5 times`

```python
## Identifies the final maximum value and 
## the no. of times the maximum value was updated

# Imports the randint function from the random module
from random import randint

# Sets the variable for the current max. value 
# to a random integer from 1 to 100
curr_max = randint(1,100)

# Displays the current maximum value
print(curr_max)

# Sets the variable for the number of updates to 0
n_update = 0

# Displays the remaining 99 random integers from 1 to 100
for i in range(99):
    num = randint(1, 100)
    # If the random number is greater than
    # the current maximum value, it will display the
    # number followed with the message "<== update"
    # Then the random number will be the
    # new current maximum value and the variable for 
    # the number of updates will be added by 1
    if num > curr_max:
        print(num, "<== Update")
        curr_max = num
        n_update += 1
    # Else, it will just display the random number
    else:
        print(num)
        
# Displays the final maximum value and the no. of times 
# the maximum value was updated.
print("The maximum value found was", curr_max)
print("The maximum value was updated", n_update, "times")
```

## Exercise 84 - Coin Flip Simulation

<p>What’s the minimum number of times you have to flip a coin before you can have
three consecutive flips that result in the same outcome (either all three are heads or
all three are tails)? What’s the maximum number of flips that might be needed? How
many flips are needed on average? In this exercise we will explore these questions
by creating a program that simulates several series of coin flips.</p>
<p>Create a program that uses Python’s random number generator to simulate flipping
a coin several times. The simulated coin should be fair, meaning that the probability
of heads is equal to the probability of tails. Your program should flip simulated
coins until either 3 consecutive heads of 3 consecutive tails occur. Display an H each
time the outcome is heads, and a T each time the outcome is tails, with all of the
outcomes for one simulation on the same line. Then display the number of flips that
were needed to reach 3 consecutive occurrences of the same outcome. When your
program is run it should perform the simulation 10 times and report the average
number of flips needed. Sample output is shown below:</p>

`H T T T (4 flips)
H H T T H T H T T H H T H T T H T T T (19 flips)
T T T (3 flips)
T H H H (4 flips)
H H H (3 flips)
T H T T H T H H T T H H T H T H H H (18 flips)
H T T H H H (6 flips)
T H T T T (5 flips)
T T H T T H T H T H H H (12 flips)
T H T T T (5 flips)
On average, 7.9 flips were needed.`

```python
## Displays the average no. of coin flips needed before
## you can have three consecutive flips that results in
## a same outcome per simulation

# Import the choices function from the random function
from random import choices

# Created a variable that simulates as a coin.
# The variable contain a list of two elements,
# "H" for head and "T" for Tail.
coin = ["H", "T"]

# Created three counters for the no. of simulations,
# no. of occurences of head, and no. of occurences
# of tail. 
sim_counter = 0
H_counter = 0
T_counter = 0

# Initialize the total no. of flips to 0
total_flips = 0

# Displays the result of the 10 coin flip simulations
while sim_counter != 10:
    # Variable for the result of the coin flip simulation
    result = choices(coin, k = 50) 
    # Created variable containing an empty list. This is where
    # the list elements of the variable "result" will be appended.
    head_tail_3 = []
    # Appends each list element from the variable "result" until the last three
    # consecutive elements are the same.
    for i in range(50):
        head_tail_3.append(result[i])
        # If the value of the list_element is "H", the H_counter is added by 1
        if result[i] == 'H':
            H_counter += 1
            # The T_counter will reset to zero if the "T" is followed by "H".
            T_counter = 0 
        # Else, the T_counter is added by 1
        else:
            # The T_counter will reset to zero if the "H" is followed by "T".
            H_counter = 0
            T_counter += 1
        # If either of the H_counter or T_counter reaches 3, it will stop appending
        # list elements from the variable "result".
        if H_counter == 3 or T_counter == 3:
            break
    
    # Variable for the number of flips per simulation
    flips = len(head_tail_3)
    
    # The no. of flips per simulation will be added to the
    # total no. of flips
    total_flips += flips
    
    # Displays each result of the coin flip simulation
    for element in head_tail_3:
        print(element, end = " ")
    print("(%d flips)" % flips)
    
    # The sim_counter will be added by 1 after 
    # executing the statements inside the "while" loop.
    sim_counter += 1

# Variable for the average no. of flips
avg_flips = total_flips/sim_counter

# Displays the average no. of flips
print("On average,",avg_flips,"flips were needed.")
```
