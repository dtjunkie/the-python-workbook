## Exercise 85: Compute the Hypotenuse

Write a function that takes the lengths of the two shorter sides of a right triangle as
its parameters. Return the hypotenuse of the triangle, computed using Pythagorean
theorem, as the function’s result. Include a main program that reads the lengths of
the shorter sides of a right triangle from the user, uses your function to compute the
length of the hypotenuse, and displays the result.  $ (23 $ $ lines) $

```python
## Computes the hypotenuse of the right triangle
# Imports the sqrt function from the math module
from math import sqrt

# @param a the length of the first side
# @param b the length of the second side
# returns the hypotenuse of the right triangle
def hypotenuse(a, b):
    c = sqrt((a ** 2) + (b ** 2))
    return c

def main():
    # Asks the user to input the length of the two sides of the right triangle
    side_a = float(input("Enter the length of the first side: "))
    side_b = float(input("Enter the length of the second side: "))
    
    # Computes and displays the hypotenuse of the triangle
    hyp = hypotenuse(side_a, side_b)
    print("The hypotenuse of the right triangle is: ", hyp, "units")

# Calls the main() function
main()
```

## Exercise 86: Taxi Fare

In a particular jurisdiction, taxi fares consist of a base fare of $\$4.00 $, plus $\$0.25 $
for every 140 meters travelled. Write a function that takes the distance travelled (in
kilometers) as its only parameter and returns the total fare as its only result. Write a
main program that demonstrates the function. $ (22 $ $ lines) $

> Hint: Taxi fares change over time. Use constants to represent the base fare and
the variable portion of the fare so that the program can be updated easily when
the rates increase.

```python
## Calculates the total taxi fare of the travel
# @param distance the distance of the travel in km
def taxi_fare(distance):
    # Set base fare to $4.00
    base_fare = 4.00
    
    # Set the additional distance to 140 m.
    add_dist = 140
    
    # Set the additional cost per every 140 m. to $0.25
    add_dist_fare = 0.25
    
    # Converts distance in km. to distance in m.
    distance_m = distance * 1000
    
    # Multiplies the additional cost per every 140 m.
    # to get the total additional cost
    multiplier = distance_m // add_dist
    
    # The total additional cost to be added to the
    # base fare
    total_add_dist_fare = add_dist_fare * multiplier
    
    # Computes and returns the total taxi fare of the travel
    total_fare = base_fare + total_add_dist_fare
    return total_fare

def main():
    # Asks the user to input his/her travelling distance
    distance = float(input("Enter your travelling distance (in km.): "))
    
    # Computes and displays the total taxi fare of the travel
    total = taxi_fare(distance)
    print("The total fare of your travel is: $%.2f" % total)

# Calls the main function
main()
```

## Exercise 87: Shipping Calculator

An online retailer provides express shipping for many of its items at a rate of $\$10.95 $
for the first item in an order, and $\$2.95 $ for each subsequent item in the same order.
Write a function that takes the number of items in the order as its only parameter.
Return the shipping charge for the order as the function’s result. Include a main
program that reads the number of items purchased from the user and displays the
shipping charge. $ (23 $ $ lines) $

```python
## Computes the shipping charge of the order
# @param items the no. of items per order
def shipping_calculator(items):
    # Base rate is $10.95 per item
    base_rate = 10.95
    
    # $2.95 per additional item
    add_item_rate = 2.95
    
    # Computes and returns the shipping charge
    shipping_charge = base_rate + (add_item_rate * (items - 1))
    return shipping_charge

def main():
    # Asks the user if how many items will be ordered for shipping
    items = int(input("How many items for shipping? "))
    
    # Computes and displays the shipping rate
    total = shipping_calculator(items)
    print("The shipping charge of your order is: $%.2f" % total)

# Calls the main() function
main()
```

## Exercise 88: Median of Three Values

Write a function that takes three numbers as parameters, and returns the median value
of those parameters as its result. Include a main program that reads three values from
the user and displays their median. $ (43 $ $ lines) $

> Hint: The median value is the middle of the three values when they are sorted
into ascending order. It can be found using if statements, or with a little bit of
mathematical creativity

```python
## Determines the median of the three numbers
## entered by the user
# @ params a,b,c the three integers
def median_three(a,b,c):
    # If b is less than a but greater than c
    # or b is greater than a but less than c
    # then, b is the median.
    if a < b < c or a > b > c:
        return b
    
    # Else, if a is greater than or less than
    # b and c, then c is the median.
    elif a > b and a > c or a < b and a < c:
        return c
    
    # Else, a is the median
    else:
        return a
    
def main():
    # Asks the user to enter three integers
    num_1 = int(input("Enter your first number: "))
    num_2 = int(input("Enter your second number: "))
    num_3 = int(input("Enter your third number: "))
    
    # Determines and displays the median of the three numbers
    median = median_three(num_1, num_2, num_3)
    print("The median of the three numbers is: ", median)

# Calls the main () function
main()
```

## Exercise 89: Convert an Integer to Its Ordinal Number

Words like $ first $, $ second $ and $ third $ are referred to as ordinal numbers. In this exercise,
you will write a function that takes an integer as its only parameter and returns a
string containing the appropriate English ordinal number as its only result. Your
function must handle the integers between 1 and 12 (inclusive). It should return an
empty string if the function is called with an argument outside of this range. Include
a main program that demonstrates your function by displaying each integer from 1
to 12 and its ordinal number. Your main program should only run when your file has
not been imported into another program. $ (47 $ $ lines) $

```python
## Converts an integer to an ordinal number
# @ param n the integer
# returns an ordinal number
def int_to_ordinal(n):
    if n == 1:
        return "first"
    elif n == 2:
        return "second"
    elif n == 3:
        return "third"
    elif n == 4:
        return "fourth"
    elif n == 5:
        return "fifth"
    elif n == 6:
        return "sixth"
    elif n == 7:
        return "seventh"
    elif n == 8:
        return "eighth"
    elif n == 9:
        return "ninth"
    elif n == 10:
        return "tenth"
    elif n == 11:
        return "eleventh"
    elif n == 12:
        return "twelfth"
    else:
        return " "

def main():
    # Asks the user to input a number
    number = int(input("Enter a number: "))
    
    # Converts and displays the ordinal number
    ordinal = int_to_ordinal(number)
    print("The ordinal of", number, "is:", ordinal)

# Calls the main() function
main()
```
